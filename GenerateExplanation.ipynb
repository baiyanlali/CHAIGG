````python
prompt = f"""
You are a game desinger.

Now given the game entities and the game rules, please match each rule using following types.

# Entities

## Entity:
Name: Player
Description: Controls movement and bounces crates.

## Entity:
Name: Crate
Description: Moves and bounces after collisions.

## Entity:
Name: Hole
Description: Disappears when a crate lands on it.

# Rules

## Rule1:
From: Player
To: Crate
Who Take Effect: Player
Effect: If the player collides with a crate, both the player and the crate step back.

## Rule2
From: Crate
To: Hole
Who Take Effect: Crate
Effect: If a crate lands directly on a hole, both the hole and the crate are removed from the level.

## Rule3
From: Crate
To: Player
Who Take Effect: Crate
Effect: If a crate hits the player, both the player and the crate step back.

## Rule4
From: Crate
To: Wall
Who Take Effect: Crate
Effect: If the crate attempts to move into a wall, the crate immediately steps back to the cell it occupied before attempting the move.

## Rule5
From: Crate
To: Wall
Effect: If the crate attempts to move into a wall, the crate immediately steps back to the cell it occupied before attempting the move.

Here is the game effect type.
""" + arguments + """
Please return result using follow format:

```json
[
  {
     "source": "EntityName",
      "target": "Entity2Name",
      "who": "The entity take the effect",
      "effect": "EffectName",
      "arguments": "Output arguments in a JSON format",
      "explain": "explaination"
  },
]
```
"""
````
